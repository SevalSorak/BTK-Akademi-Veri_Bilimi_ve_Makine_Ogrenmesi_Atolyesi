```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmis_20220204.csv")
```

```python
veri.head()
```

## Saçılım Grafiği

```python
sns.scatterplot(x = "boy", y = "kilo", data = veri)
plt.title("Boy ve Ağırlık İlişkisi")
sns.set_style("ticks")
plt.show()
```

```python
sns.scatterplot(x = "boy", y = "kilo", hue = "madalya", data = veri, palette=["#ffd700", "#CD7F32", "#c0c0c0"])
plt.title("Madalyaya Göre Boy ve Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x = "boy", y = "kilo", hue = "cinsiyet", style = "madalya", data = veri, 
                palette=["#4F5BD5", "#FFAAA5"], markers=["o", "v", "*"],
                size = "yas", sizes=(15,200))
plt.title("Madalyalı Sporcuların Boy ve Ağırlık Dağılımı")
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.show()
```

```python
sns.regplot(x = "boy", y = "kilo", data = veri, marker="P", scatter_kws={"alpha":0.2}, color="#D62976")
plt.title("Boy ve Ağırlık Dağılımı")
plt.show()
```

## Çizgi Grafiği

```python
sns.lineplot(x = "boy", y = "kilo", data = veri, color = "#A0522D")
plt.title("Boy ve Kilo")
plt.show()
```

```python
sns.lineplot(x = "boy", y = "kilo", data = veri, hue = "cinsiyet", palette = ["#243447", "#C51F5D"])
plt.title("Cinsiyete göre Boy ve Kilo")
plt.show()
```

## Histogram

```python
sns.displot(veri, x = "kilo", color="#FF48C4")
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

```python
sns.displot(veri, x = "kilo", hue = "cinsiyet", palette=["#2BD1FC", "#F3EA5F"])
plt.ylabel("Frekans")
plt.title("Cinsiyete göre Ağırlık Histogramı")
plt.show()
```

```python
sns.displot(veri, x = "kilo", col = "cinsiyet", multiple = "dodge", color = "#C04DF9")
plt.show()
```

```python
sns.displot(veri, x = "kilo", y = "boy", kind = "kde", rug = True, color = "#FF3F3F")
plt.ylabel("Boy")
plt.xlabel("Kilo")
plt.title("Ağırlık - Boy Histogramı")
plt.show()
```

```python
sns.displot(veri, x = "kilo", y = "boy", hue = "cinsiyet", kind = "kde", rug = True, palette=["#7BB3FF", "#E86AF0"])
plt.ylabel("Boy")
plt.xlabel("Kilo")
plt.title("Cinsiyete göre Ağırlık - Boy Histogramı")
plt.show()
```

```python
sns.kdeplot(data = veri, x = "kilo", y = "boy", hue = "cinsiyet", fill = True, palette=["#0BD3D3", "#F890E7"])
plt.ylabel("Boy")
plt.xlabel("Kilo")
plt.title("Cinsiyete göre Ağırlık - Boy Histogramı")
plt.show()
```

## Çubuk Grafiği

```python
sns.barplot(x = "madalya", y = "boy", data = veri, palette=["#ffd700", "#967444", "#c0c0c0"])
plt.title("Madalyalara göre Boy Grafikleri")
plt.show()
```

```python
sns.barplot(x = "madalya", y = "boy", data = veri, hue = "cinsiyet", palette=["#1B85B8", "#559E83"])
plt.title("Madalyalara göre Boy Grafikleri")
plt.show()
```

```python
sns.catplot(x = "madalya", y = "yas", data = veri, hue = "cinsiyet", palette=["#18FFF9", "#FF0000"],
            kind = "bar", col = "sezon", height = 4, aspect = .9)
plt.show()
```

## Kutu Grafiği

```python
sns.boxplot(x = "sezon", y = "boy", data = veri, palette = ["#065535", "#ff80ed"])
plt.show()
```

```python
sns.boxplot(x = "sezon", y = "boy", hue = "cinsiyet", data = veri, palette = ["#000000", "#ffc0cb"])
plt.show()
```

```python
veri_gecici = veri.loc[:,["yas", "kilo", "boy"]]
sns.boxplot(data = veri_gecici, orient="h", palette=["#ffe4e1", "#133337", "#ffc0cb"])
plt.show()
```

```python
sns.catplot(x = "sezon", y = "boy", hue = "cinsiyet", col = "madalya", kind = "box",
            palette=["#008080", "#ff0000"], data = veri, height=4, aspect=.7)
plt.show()
```

## Isı Haritası

```python
sns.heatmap(veri.corr(), annot=True, linewidths=0.5, fmt=".1f", cmap="mako")
plt.show()
```

## Keman Grafiği

```python
sns.violinplot(x = "sezon", y = "boy", data = veri, palette=["#e6e6fa", "#ffd700"])
plt.show()
```

```python
sns.violinplot(x = "sezon", y = "boy", hue = "cinsiyet", data = veri, palette=["#00ffff", "#ffa500"])
plt.show()
```

```python
sns.violinplot(x = "sezon", y = "boy", hue = "cinsiyet", data = veri, split=True, palette=["#00ffff", "#ffa500"])
plt.show()
```

```python
sns.catplot(x = "sezon", y = "boy", hue = "cinsiyet", col="madalya",
            data = veri, kind="violin", split=True, height=4, aspect=.7,
            palette=["#0000ff", "#ff7373"])
plt.show()
```

## Ortaklık Grafiği

```python
sns.jointplot(data = veri, x = "kilo", y = "boy", hue = "sezon", kind = "kde",
              palette=["#c6e2ff", "#40e0d0"])
plt.show()
```

```python
g = sns.JointGrid(data = veri, x = "kilo", y = "boy")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

## Çift / Eş Grafiği

```python
sns.pairplot(veri)
plt.show()
```

```python
g = sns.PairGrid(veri)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill = True)
g.map_diag(sns.histplot, kde = True)
```
