```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
veri = {"a" : np.arange(50),
        "c" : np.random.randint(0,50,50),
        "d" : np.random.randn(50)}
veri["b"] = veri["a"] + (10 * np.random.randn(50))
veri["d"] = np.abs(veri["d"]) * 100
```

```python
plt.scatter("a", "b", c = "c", s = "d", data = veri)
plt.xlabel("a sütunundaki veriler")
plt.ylabel("b sütunundaki veriler")
plt.show()
```

```python
isimler = ["Ahmet", "Mehmet", "Ayşe", "Fatma"]
degerler = [5, 25, 50, 100]

plt.figure(figsize=(10,3))

plt.subplot(131) # 1 satır, 3 sütun olsun. 1.si
plt.bar(isimler, degerler)

plt.subplot(132) # 1 satırlık, 3 sütunluk  yerin 2.si
plt.scatter(isimler, degerler)

plt.subplot(133) # 1 satırlık, 3 sütunluk  yerin 3.sü
plt.plot(isimler, degerler)

plt.suptitle("Birçok Farklı Grafik")
plt.show()
```

```python
t = np.arange(0,5,0.2)
plt.plot(t, t, "c-", linewidth=5)
```

```python
t = np.arange(0,5,0.1)
plt.plot(t, (np.exp(-t) * np.cos(2 * np.pi * t)), "#FF6600", t, (np.exp(-t) * np.cos(2 * np.pi * t)), "mo")
plt.axis([0,10,-1,1])
plt.show()
```

```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))
```

```python
plt.figure(figsize=(9,6))
n, bins, patches = plt.hist(x, 100, density=1, facecolor="#FFAD60", alpha=0.40)
n, bins, patches = plt.hist(y, 100, density=1, facecolor="#7289DA", alpha=0.40)

plt.xlabel("Veriler")
plt.ylabel("Olasılıklar")
plt.title(r"Verilerin Histogramı: $\mu_1=100,\ \mu_2=110$")
plt.annotate("Gauss eğrisinin tepesi", xy=(110,0.04), xytext=(130,0.045),
             arrowprops=dict(facecolor="#800080", shrink=0.1))
plt.text(72,0.025, r"$\mu_1=100,\ \sigma_1=15$")
plt.text(120,0.033, r"$\mu_2=110,\ \sigma_2=10$")
plt.axis([40,160,0,0.05])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0,10,0.01)
y = np.exp(x)
```

```python
plt.figure()

plt.subplot(121)
plt.plot(x,y, color="#FF9966")
plt.yscale("linear")
plt.title("Lineer(Doğrusal)")
plt.grid(True)

plt.subplot(122)
plt.plot(x,y, color="#FF9966")
plt.yscale("log")
plt.title("Logaritmik")
plt.grid(True)

plt.show()
```
